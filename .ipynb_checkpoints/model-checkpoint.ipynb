```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("laptops.csv")
df
```

```python
df.drop(["id","model","comments"],axis="columns",inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df = df.fillna({"graphic_card_size":0,"weight":df.weight.mean(),"clock_speed":df.clock_speed.mean()})
```

```python
df.dropna(inplace=True)
```

```python
df1 = df[df.processor_brand=="intel"]
```

```python
df1.drop("processor_brand",axis="columns",inplace=True)
df1
```

```python
df1.processor_model.value_counts()
```

```python
def pro(x):
    x=x.lower()
    y=x.split(" ")[0]
    if y.startswith("i"):
        return y
    else:
        return "other"
```

```python
df1.processor_model = df1.processor_model.apply(pro)
```

```python
df1
```

```python
sns.countplot(x='processor_model',data=df1)
```

```python
processor_model_encoder = LabelEncoder()
df1.processor_model = processor_model_encoder.fit_transform(df1.processor_model)
processor_model_encoder.classes_
```

```python
os_counts = df1['os'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
os_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Operating System')
plt.ylabel('Frequency')
plt.title('Distribution of Operating Systems')
plt.xticks(rotation=0)
plt.show()
```

```python
df1.os.value_counts()
```

```python
df2=df1[df1.os!="windows,linux"]
```

```python
os_encoder = LabelEncoder()
df2.os = os_encoder.fit_transform(df2.os)
os_encoder.classes_
```

```python
df2.weight.dtype
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df2.drop("price",axis="columns"),df2.price,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
l = LinearRegression()
l.fit(x_train,y_train)
print("Testing data r squared score: ", l.score(x_test,y_test))
print("Training data r squared score: ",l.score(x_train,y_train))
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split

# Assuming X_train, X_test, y_train, y_test are already defined
# Train Lasso Regression model
lasso_reg = Lasso(alpha=0.1)  # You can adjust the alpha hyperparameter
lasso_reg.fit(x_train, y_train)

# Evaluate Lasso Regression model
lasso_reg_score_test = lasso_reg.score(x_test, y_test)
lasso_reg_score_train = lasso_reg.score(x_train, y_train)

print("Lasso Regression R^2 Score (Test):", lasso_reg_score_test)
print("Lasso Regression R^2 Score (Train):", lasso_reg_score_train)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
import joblib

# Assuming X_train, X_test, y_train, y_test are already defined
# Train Ridge Regression model
ridge_reg = Ridge(alpha=1.0)  # You can adjust the alpha hyperparameter
ridge_reg.fit(x_train, y_train)

# Evaluate Ridge Regression model
ridge_reg_score_test = ridge_reg.score(x_test, y_test)
ridge_reg_score_train = ridge_reg.score(x_train, y_train)

print("Ridge Regression R^2 Score (Test):", ridge_reg_score_test)
print("Ridge Regression R^2 Score (Train):", ridge_reg_score_train)

# Save the Ridge Regression model
joblib.dump(ridge_reg, "model.pkl")
```
